## Visualizations Data

```python
#Import modules
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Connect sql server and homework database
engine = create_engine('postgresql://postgres:postgres@localhost:5432/sql_homework')
conn = engine.connect()
```

```python
#Use pandas to query the sql salary data for our visuals
salary = pd.read_sql('SELECT * FROM "Salaries"' , conn)
salary.head()
```

```python
#Use pandas to query the sql employee data for our visuals
employee = pd.read_sql('SELECT * FROM "Employees"' , conn)
employee.head()
```

```python
#Use pandas to query the sql titles data for our visuals
title = pd.read_sql('SELECT * FROM "Titles"' , conn)
title.head()
```

```python
#Merge the 2 datasets together to be able to create the histogram correctly
combo1_db = employee.merge(salary , on = "emp_no")
combo1_db
```

```python
#Then we take the merged db and merge it again on title to be able to use it for the last visual
final_combo_db = combo1_db.merge(title , left_on = "emp_title_id" , right_on = "title_id")
final_combo_db
```

## Visualizations

```python
#Prep the above db for the first visualization
histogram_data = final_combo_db[["salary" , "title"]]

#Groupby is the easiest way to create our visualization dataset
histogram_data.groupby("title")["salary"].mean().round()

#Create a histogram to visualize the most common salary ranges for employees
sns.set(style="darkgrid")
plt.figure(figsize = (13,11))

#Adding colors to each histogram bar and making it look nicer
n, bins, patches = plt.hist(histogram_data["salary"], bins=9, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5)

#Must be integer for this to work
n = n.astype('int')

#Need this to be able to give each bar a color from a palette
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.copper(n[i]/max(n)))

#Add the labels and titles
plt.title("Salary Distribution" , size =20)
plt.xlabel("Salaries ($)" , size = 15)
plt.ylabel("Salary Frequency" , size = 15)
plt.savefig("../images/employee_salary_histogram.PNG")
plt.show()
```

```python
#Create a bar chart of average salary by title.
#Use the above variable to plot the barchart
bar_chart_data = histogram_data.groupby("title")["salary"].mean().round()

#Formatting and colors
palette = list((sns.color_palette("copper", as_cmap=False)))
sns.set(style="darkgrid")
plt.figure(figsize = (13,11))

#Plotting the bar chart and adding labels
plt.bar(bar_chart_data.index,bar_chart_data.values , color = palette)
plt.title("Salaries by Title", size = 20)
plt.xlabel("Emoployee Title", size = 15)
plt.ylabel("Salaries ($)", size = 15)
plt.xticks(rotation =25)
plt.savefig("../images/employee_salary_barchart.PNG")
plt.show()
```
